```python
merge_df = pd.read_excel("data/路虎口碑所有用户.xlsx")
# create new column post_count, excellent_post_count
# define extraction function
def extract_excellent_post_count(row):
    import re
    return int(re.findall("\d+",row["post_count"].split("/")[0])[0])
def extract_post_count(row):
    import re
    return int(re.findall("\d+",row["post_count"].split("/")[1])[0])

merge_df["excellent_post_count"]=merge_df.apply (lambda row: extract_excellent_post_count(row),axis=1)
merge_df["all_post_count"]=merge_df.apply (lambda row: extract_post_count(row),axis=1)
merge_df.drop(columns=["post_count"],inplace=True)
print(merge_df.shape)
merge_df.head()
```

### “关注的车”

```python
def extract_car_info(row):
    s = row["car_liked"].strip().split("：")
    if len(s)>1:
        return s[1]
    else:
        return ''
merge_df["car_like"] = merge_df.apply(lambda row:extract_car_info(row),axis=1)
```

```python
merge_df = merge_df.drop(columns=['car_liked'])
merge_df.head()
```
